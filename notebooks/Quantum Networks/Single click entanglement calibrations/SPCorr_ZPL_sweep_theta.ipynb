```python
### calculate some theta vlaues

A = 0.510846
f = 0.576345
x0 = -0.064609
def get_theta(x):
    
    population = (1-A)+ A*np.cos(2*np.pi*f*(x-x0))
    sintheta = np.sqrt(population)
    return np.arcsin(sintheta)/np.pi

print get_theta(0.57)
print get_theta(0.2)
```
