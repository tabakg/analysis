```python
num_bins = 20

bins = range(num_bins)
y, binedges =  np.histogram(Blocks, bins)
x = range(num_bins-1)

save_num = 2
begin = True
indices = []

for i,j in enumerate(y):
    if j == 0:
        if i != save_num + 1:
            if begin:
                indices.append(i+1)
                begin = False
                save_nume = i
            else:
                indices.append(i)
                begin = True
                save_num = i

for i in range(len(indices)/2):
    if i == 0:
        start = indices[(2*i)]
        stop = indices[(2*i)+1]
        y1 = y[start:stop]
        x1 = x[start:stop]
    else:
        start = indices[(2*i)]
        stop = indices[(2*i)+1]
        y1 = np.hstack((y1,y[start:stop]))
        x1.extend(x[start:stop])


fig, ax = plt.subplots()

ax.plot(x1,y1,'bo-')
ax.set_xlabel('Block Length (in bits)')
ax.set_ylabel('Occurence')
ax.set_xlim(0,17)

ax.set_yscale('log')

plt.draw()
plt.show()
```
