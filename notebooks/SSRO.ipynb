```python
folder = r'N:\tnw\NS\qt\Diamond\Autobackup\lt3\data\20150226\145811_AdwinSSRO_SSROCalibration_Pippin_SIL1'
save_folder = r'H:\My Documents\processed data'
a = ssro.SSROAnalysis(folder) 
aa=a.analysis_h5data()

#Plot settings
twinplot=False
ms1_logplot=True
last_ro_bin=8
ro_counts=[]
xmax = 12
bins = np.arange(100)
tick_stepsize=1
plot_colors = ['b','b']

gr=scipy.constants.golden_ratio
if twinplot:
    fig,ax1 = plt.subplots(1,figsize = (gr*4,4))
    ax2=ax1.twinx()
else:
    fig,(ax1,ax2) = plt.subplots(1,2,figsize = (2*gr*4,4))
for n,ms,ax,c in zip(['$m_s=0$','$m_s=\pm 1$'],['ms0', 'ms1'],[ax2,ax1],plot_colors):
    a.get_run(ms)
    ro_counts.append(a.ro_counts)
   
    cpsh = np.sum(a.ro_counts[:,0:last_ro_bin], axis=1)
    mean_cpsh = np.sum(cpsh)/float(a.reps)
    pzero = np.sum(cpsh==0)/float(a.reps)
    h,b = np.histogram(cpsh, bins=bins , normed=True)
    ax.bar(bins[:-1]-0.5, h, width=1,color=c, alpha=(0.5 if twinplot else 1), log=(ms=='ms1' and ms1_logplot))
    ax.set_xlim(-0.5,xmax-0.5)
    ax.set_xticks(np.arange(0,xmax,tick_stepsize))
    ax.set_xlabel('cts/shot')
    ax.text(0.93*xmax,ax.get_ylim()[1]*(1-0.1*twinplot),'\np(0 counts) = {:.2f}%'.format(pzero*100), verticalalignment='top',horizontalalignment='right', color=c)
    
    if twinplot:
        ax.set_ylabel('Probability after ' + n + ' initialisation', color=c)
    else:
        ax.set_ylabel('Probability')
        ax.set_title('Intitialisation in ' +n)    
mean_fidelity = aa['fidelity']['mean_fidelity'].value[last_ro_bin]
u_mean_fidelity = aa['fidelity']['mean_fidelity_err'].value[last_ro_bin]
fig.suptitle('Detected clicks per {:d} us readout\n (average fidelity: ${:.1f}\%\pm{:.2f}\%$)'.format(last_ro_bin-1,mean_fidelity*100,u_mean_fidelity*100),verticalalignment='bottom', fontsize = 14)
fig.savefig(os.path.join(save_folder,a.name+('_twinplot' if twinplot else '')+'.pdf'),bbox_inches='tight')
```
