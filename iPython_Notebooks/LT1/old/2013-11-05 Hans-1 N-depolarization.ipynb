## N-spin depolarization by pumping on E_12

```python
def nspinflips(folder):
    a = mbi.MBIAnalysis(folder)
    a.get_sweep_pts()
    a.get_readout_results('adwindata')
    a.get_electron_ROC()
    a.finish()
    
    f = h5py.File(os.path.join(folder, 'analysis.hdf5'), 'r')
    x = f['/readout_result/sweep_pts'].value.reshape(-1)
    y = f['/readout_result/p0'].value.reshape(-1)
    u_y = f['/readout_result/u_p0'].value.reshape(-1)
    f.close()

    A = fit.Parameter(0.95, 'A')
    p = fit.Parameter(0.001, 'p')
    o = fit.Parameter(0, 'o')
    p0 = [A, p, o]
    fitfunc_str = 'A * 1/6 (2 + (1-3p)^n + 3(1-p)^n) + o'

    def ff(x):
        return A()/6. * (2. + (1-3*p())**x + 3*(1-p())**x) + o()

    fitres = fit.fit1d(x, y, None, p0=p0, fitfunc=ff, fitfunc_str=fitfunc_str,
                       do_print=True, ret=True)
    fitinfo = fitres['fitfunc_str'] + '\n' + fit.str_fit_params(fitres)

    fig, ax = subplots(1,1)
    ax.errorbar(x, y, yerr=u_y, fmt='o')
    fitx = linspace(0,2500,201)
    ax.plot(fitx, ff(fitx), lw=2)

    # ax.set_xscale('log')
    ax.set_xlabel('Number of electron spin flips')
    ax.set_ylabel('(Unnormalized) population of $m_I = -1$')
    ax.text(1000, 0.9, fitinfo, va='top', ha='left', color='k', size='x-small',
             bbox = {'facecolor' : 'white', 'alpha' : 0.5})

    ax.set_title('Nuclear spin depolarization by optical spin pumping')
    
```
