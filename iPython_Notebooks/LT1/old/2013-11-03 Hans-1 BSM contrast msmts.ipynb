```python
def BSM_contrast(folder):
    a = mbi.MBIAnalysis(folder)
    a.get_readout_results('adwindata')
    a.get_sweep_pts()

    reps = a.g.attrs['reps_per_ROsequence']
    N_RO_reps = a.adgrp.attrs['N_RO_repetitions']
    pts = a.adgrp.attrs['sweep_length']
    readouts = a.adgrp.attrs['nr_of_ROsequences']
    results = a.adgrp['ssro_results'].value.reshape((-1,pts,readouts))
    x = a.sweep_pts.reshape(-1)
    
    fig = a.default_fig(figsize=(6,4))
    ax = a.default_ax(fig)

    a.finish()

    # electron results are the first outcome
    e_results = results[:,:,0]

    # N results are determined whether we found at least one click in a following readout
    # (filtering on a max. no. of RO reps is not in right now, but easy to implement by adapting the range)
    N_results = results[:,:,1:].sum(axis=-1)
    N_results[N_results>1] = 1

    # nitrogen goes first; note that the MBI program saves whether we got a click, not m_s/m_I 
    corr_names = ['00', '01', '10', '11']
    corrs = 2**1 * abs(1-N_results) + 2**0 * abs(1-e_results)

    # compute the probability for all correlations
    p_corr = empty((4,pts))
    u_p_corr = empty((4,pts))
    for i in range(4):
        for j in range(pts):
            n = float(len(corrs[:,j][corrs[:,j]==i]))
            p_corr[i,j] = n/reps
            u_p_corr[i,j] = sqrt((n/reps)*(1.-n/reps)/reps)
        
    for i,c in enumerate(corr_names):
        ax.errorbar(x, p_corr[i,:], yerr=u_p_corr[i,:], fmt='o', 
                    label=c)
    
    ax.set_xlim(-10, 370)
    ax.set_ylim(-0.05,1.05)
    ax.set_xlabel('Phase')
    ax.set_ylabel('Probability')
    ax.legend(loc='best')
```
